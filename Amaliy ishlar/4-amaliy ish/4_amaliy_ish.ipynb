```python
df2 = pd.read_csv("/content/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df2.head()
```

```python
df2['month'] = [i.month for i in df2.index]
df2['year'] = [i.year for i in df2.index]
df2.head()
```

```python
df2.groupby(['month','year']).max()
```

```python
df_m = df2.groupby(['month','year']).max().unstack(level=0)
df_m.head()
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(data=df_m, cmap="Blues", linewidths=1, square=True)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels)
plt.show()
```
